```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Trade_d=pd.read_csv("Trade.csv")
Export_d=pd.read_csv("Export_2.csv")
```

```python
Trade_d.drop("cname", inplace=True, axis=1)
```

```python
Trade_d.drop("export", inplace=True, axis=1)
```

```python
Trade_d.shape
```

```python
Trade_d
```

```python
Data_1 = Trade_d.loc[Trade_d['year'] == 2004]
Data_1=Data_1.merge(Export_d[['Country Code', '2004']], left_on='ccode', right_on='Country Code', how='left')
Data_1.drop('Country Code', inplace=True, axis=1)
new_column_names = ['year','ccode','Physical_infrastructure','ICT','Business','Border_Transportation','Export']
Data_1.columns=new_column_names
Data_1

Data_2 = Trade_d.loc[Trade_d['year'] == 2005]
Data_2=Data_2.merge(Export_d[['Country Code', '2005']], left_on='ccode', right_on='Country Code', how='left')
Data_2.drop('Country Code', inplace=True, axis=1)
new_column_names = ['year','ccode','Physical_infrastructure','ICT','Business','Border_Transportation','Export']
Data_2.columns=new_column_names
Data_2

Data_3 = Trade_d.loc[Trade_d['year'] == 2006]
Data_3=Data_3.merge(Export_d[['Country Code', '2006']], left_on='ccode', right_on='Country Code', how='left')
Data_3.drop('Country Code', inplace=True, axis=1)
new_column_names = ['year','ccode','Physical_infrastructure','ICT','Business','Border_Transportation','Export']
Data_3.columns=new_column_names
Data_3

Data_4 = Trade_d.loc[Trade_d['year'] == 2007]
Data_4=Data_4.merge(Export_d[['Country Code', '2007']], left_on='ccode', right_on='Country Code', how='left')
Data_4.drop('Country Code', inplace=True, axis=1)
new_column_names = ['year','ccode','Physical_infrastructure','ICT','Business','Border_Transportation','Export']
Data_4.columns=new_column_names
Data_4

Trade_d = pd.concat([Data_1, Data_2, Data_3, Data_4], ignore_index=True)
Trade_d
```

```python
Trade_d.isnull().sum()*100/Trade_d.shape[0]
```

```python
Trade_d.dropna(inplace = True)
Trade_d.isnull().sum()
```

```python
Trade_d.shape
```

```python
Trade_d.to_csv('merged_data.csv', index=False)
```

```python
Trade_d=pd.read_csv("merged_data_3.csv")
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()

Trade_d["Export"]=scaler.fit_transform(Trade_d[["Export"]])
```

```python
condition = (Trade_d['year'] == 2005)
Trade_d['2005'] = (condition).astype(int)

condition = (Trade_d['year'] == 2006)
Trade_d['2006'] = (condition).astype(int)

condition = (Trade_d['year'] == 2007)
Trade_d['2007'] = (condition).astype(int)
```

```python
Data=Trade_d
```

```python
dimin = Data.groupby('ccode').sum().reset_index()
dimin
```

```python
new_column_names = ['ccode','year','Physical_infrastructure_bar','ICT_bar','Business_bar','Border_Transportation_bar','Export_bar','2004','2005','2006']
dimin.columns=new_column_names
dimin
```

```python
Trade_d=Trade_d.merge(dimin[['ccode', 'Physical_infrastructure_bar','ICT_bar','Business_bar','Border_Transportation_bar','Export_bar']], left_on='ccode', right_on='ccode', how='left')
Trade_d
```

```python
Trade_d['Delta_Physical_infrastructure']=Trade_d['Physical_infrastructure']-(Trade_d['Physical_infrastructure_bar']/4)
```

```python
Trade_d['Delta_ICT']=Trade_d['ICT']-(Trade_d['ICT_bar']/4)
```

```python
Trade_d['Delta_Business']=Trade_d['Business']-(Trade_d['Business_bar']/4)
```

```python
Trade_d['Delta_Border_Transportation']=Trade_d['Border_Transportation']-(Trade_d['Border_Transportation_bar']/4)
```

```python
Trade_d['Delta_Export']=Trade_d['Export']-(Trade_d['Export_bar']/4)
Trade_d.head(-1)
```

```python
Trade_d.to_csv('merged_data_f.csv', index=False)
```

```python

import statsmodels.api as sm

X_train1 = sm.add_constant(X_train)
model = sm.OLS(Y_train, X_train1).fit(cov_type='HC3')

#X_test1 = sm.add_constant(X_test)
X_test1 = sm.add_constant(X_train)
predictions = model.predict(X_test1) 

print_model = model.summary()
print(print_model)
```
